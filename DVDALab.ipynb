```python
df = pd.read_csv('datasets/winequality-red.csv')
```

```python
df.head()
```

```python
qualityCount=df['quality'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(qualityCount.index,qualityCount,color = 'Maroon')
plt.title('Count Plot of Quality of wine')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
#density Plot
sns.set_style('darkgrid')
```

```python
numericalColumn = df.select_dtypes(include=['int64','float64']).columns
```

```python
#kde ဆိုတာ အပေါ်မှာ ပေါ်နေတဲ့ မျဥ်းေကြာင်းကိုေပြာတာဖြစ် 
plt.figure(figsize=(14,len(numericalColumn)*5))
for idx , feature in enumerate(numericalColumn,1):
    plt.subplot(len(numericalColumn),2,idx)
    sns.histplot(df[feature],kde=True)
    plt.title(f"{feature} | Skewness:{round (df[feature].skew(),2)}")
    
```

```python
plt.figure(figsize=(14,len(numericalColumn)*5))
for idx , feature in enumerate(numericalColumn,1):
    plt.subplot(len(numericalColumn),2,idx)
    sns.histplot(df[feature],kde=True)
    plt.title(f"{feature} | Skewness:{round (df[feature].skew(),2)}")
plt.tight_layout()
plt.show()
```
