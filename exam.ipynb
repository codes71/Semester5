```python
df=pd.read_csv('datasets/winequality-red.csv')
numColumn=df.select_dtypes(include=['int64','float64']).columns
plt.figure(figsize=(14,len(numColumn)*3))
sns.set_style('darkgrid')
for idx , feature in enumerate(numColumn,1):
    plt.subplot(len(numColumn),2,idx)
    sns.histplot(df[feature],kde=True)
    plt.title(f"{feature} | Skewness:{round (df[feature].skew(),2)}")
    plt.tight_layout()
    
```

```python
sns.boxplot(x='quality',y='alcohol',data = df)
```
